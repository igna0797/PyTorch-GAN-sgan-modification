###Crea copia de las imagenes y les agrega lineas

```python
import sys
import os
import torch
from torchvision.io import read_image
from torchvision.io.image import ImageReadMode
from torchvision.utils import save_image
from torchvision.transforms import Normalize, ToTensor
import subprocess
import numpy as np
from PIL import Image

from utils import add_lines

n_lineas = "1"
Random = "False"
# Set the paths
py_file_location = "/content/PyTorch-GAN-sgan-modification/implementations/sgan/"
sys.path.append(os.path.abspath(py_file_location))
images_path= f'/content/PyTorch-GAN-sgan-modification/trainings/n-lineas_{n_lineas}_Random_{Random}/images'
output_path = f'/content/PyTorch-GAN-sgan-modification/trainings/n-lineas_{n_lineas}_Random_{Random}/imagesWithLines'
os.makedirs(output_path, exist_ok=True)
# Iterate over the seeds and process images
seeds = range(1, 25)
for seed in seeds:

    # Construct image name and path
    image_name = f"seed_{seed:.3f}.png"
    image_path = os.path.join(images_path, image_name)

    # Check if the image exists
    if not os.path.exists(image_path):
        print(f"Image not found: {image_path}")
        continue

    # Read the image and make it a tensor
    image = Image.open(image_path).convert('L')
    tensor = ToTensor()
    image_tensor= tensor(image)

    # Add lines to the image tensor
    image_tensor_with_lines = add_lines(image_tensor, max_amount_lines=4, random_amount_lines=False)

    # Save the modified image tensor
    save_path = os.path.join(output_path, f"seed_{seed}_lines.png")
    save_image(image_tensor_with_lines, save_path, normalize=False)

```

###Recopila todas las imagenes en un colage

```python
from PIL import Image
import os

n_lineas_list = ["1","3","5"]
Random_list = ["False","False","True"]
for (n_lineas , Random) in zip(n_lineas_list,Random_list):
  # Path to the directory containing your images
  #input_dir = "/content/drive/MyDrive/Redes neuronales/Monografia/images"
  input_dir = f"/content/PyTorch-GAN-sgan-modification/trainings/n-lineas_{n_lineas}_Random_{Random}/images"
  os.makedirs(input_dir, exist_ok=True)
  output_file = f"n-lineas:{n_lineas}_Random:{Random}.png"  # Output file name

  # List of supported image file extensions
  image_extensions = [".jpg", ".jpeg", ".png", ".gif", ".bmp"]

  # Initialize a list to store images
  images = []

  # Iterate through the files in the directory
  for filename in os.listdir(input_dir):
      if any(filename.lower().endswith(ext) for ext in image_extensions):
          # Open and append the image to the list
          img = Image.open(os.path.join(input_dir, filename))
          images.append(img)

  # Determine the number of rows and columns in the grid
  num_images = len(images)
  num_cols = int(num_images**0.5)
  num_rows = (num_images + num_cols - 1) // num_cols

  # Calculate the dimensions of the output image
  image_width, image_height = images[0].size
  grid_width = image_width * num_cols
  grid_height = image_height * num_rows

  # Create a new blank image with the size of the grid
  grid_image = Image.new("RGB", (grid_width, grid_height))

  # Paste each image onto the grid
  for i in range(num_images):
      row = i // num_cols
      col = i % num_cols
      grid_image.paste(images[i], (col * image_width, row * image_height))

  # Save the grid image
  grid_image.save(input_dir+"/"+output_file)

  print(f"Grid image saved in {input_dir}/{output_file}")
```

# Graficar el training

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Define the file paths
filename1 = "/content/drive/MyDrive/Redes neuronales/Monografia/n-lineas:0_Random:False/training/loss_data.csv"
filename2 = "/content/drive/MyDrive/Redes neuronales/Monografia/n-lineas:3_Random:False/training/loss_data.csv"
filename3 = "/content/drive/MyDrive/Redes neuronales/Monografia/n-lineas:5_Random:True/training/loss_data.csv"

plotNames= ["1 linea de ruido", "3 lineas de ruido" , "5 lineas de ruido variable"]

# List of file paths
csv_files = [filename1, filename2, filename3]

dfs = []
means_list = []
variances_list = []
accuracies_list = []

for file in csv_files:
    df = pd.read_csv(file)
    dfs.append(df)

    means = []
    variances = []
    accuracies = []

    for epoch in df["Epoch"].unique():
        epoch_data = df[df["Epoch"] == epoch]

        discriminator_loss_mean = epoch_data["Discriminator Loss"].mean()
        generator_loss_mean = epoch_data["Generator Loss"].mean()

        # Calculate variance for loss
        discriminator_loss_variance = epoch_data["Discriminator Loss"].var()
        generator_loss_variance = epoch_data["Generator Loss"].var()

        # Calculate mean and variance for accuracy
        discriminator_accuracy_mean = epoch_data["Discriminator Accuarcy"].mean()
        discriminator_accuracy_variance = epoch_data["Discriminator Accuarcy"].var()

        means.append([discriminator_loss_mean, generator_loss_mean, discriminator_accuracy_mean])
        variances.append([discriminator_loss_variance, generator_loss_variance, discriminator_accuracy_variance])
        accuracies.append(discriminator_accuracy_mean)

    means_list.append(means)
    variances_list.append(variances)
    accuracies_list.append(accuracies)

# Create separate plots for each file and calculate mean, variance, and accuracy
figs = []
axes = []

for i, (means, variances, accuracies) in enumerate(zip(means_list, variances_list, accuracies_list)):
    fig, ax = plt.subplots(1, 3, figsize=(25, 6))
    figs.append(fig)
    axes.append(ax)

    # Plot mean, variance, and accuracy for specific columns in each file
    axes[i][0].plot(dfs[i]["Epoch"].unique(), [mean[0] for mean in means], label=f"Discriminator Loss Mean and Variance - " + plotNames[i])
    axes[i][0].fill_between(
        dfs[i]["Epoch"].unique(),
        [mean[0] - var[0] for mean, var in zip(means, variances)],
        [mean[0] + var[0] for mean, var in zip(means, variances)],
        alpha=0.2,
    )
    #axes[i][ 0].set_title(f"Discriminator Loss Mean and Variance - "+ plotNames[i])
    axes[i][ 0].set_xlabel("Epoch")
    axes[i][ 0].set_ylabel("Loss")
    axes[i][ 0].legend()

    axes[i][ 1].plot(dfs[i]["Epoch"].unique(), [mean[1] for mean in means], label=f"Generator Loss Mean - " + plotNames[i])
    axes[i][ 1].fill_between(
        dfs[i]["Epoch"].unique(),
        [mean[1] - var[1] for mean, var in zip(means, variances)],
        [mean[1] + var[1] for mean, var in zip(means, variances)],
        alpha=0.2,
    )
    #axes[i][ 1].set_title(f"Generator Loss Mean and Variance ")
    axes[i][ 1].set_xlabel("Epoch")
    axes[i][ 1].set_ylabel("Loss")
    axes[i][ 1].legend()

    axes[i][ 2].plot(dfs[i]["Epoch"].unique(), accuracies, label=f"Discriminator Accuracy Mean - " + plotNames[i])
    axes[i][ 2].fill_between(
        dfs[i]["Epoch"].unique(),
        [mean[2] - var[2] for mean, var in zip(means, variances)],
        [mean[2] + var[2] for mean, var in zip(means, variances)],
        alpha=0.2,
    )
    #axes[i][ 2].set_title(f"Discriminator Accuracy Mean and Variance ")
    axes[i][ 2].set_xlabel("Epoch")
    axes[i][ 2].set_ylabel("Accuracy")
    axes[i][ 2].legend()
   # plt.savefig(f"/content/drive/MyDrive/Redes neuronales/Monografia/Individual-Training-plot{i+1}.png")

# Create a combined plot for all files showing means, variances, and accuracy
fig_combined, ax_combined = plt.subplots(3, 1, figsize=(10, 10))

# Lists to store mean and accuracy values for specific columns in the combined data
means_combined = []
accuracies_combined = []

for i, (means, variances, accuracies) in enumerate(zip(means_list, variances_list, accuracies_list)):
    # Plot mean and accuracy for specific columns in the combined data
    ax_combined[0].plot(dfs[i]["Epoch"].unique(), [mean[0] for mean in means], label=f"Discriminator Loss Mean - " + plotNames[i])

    ax_combined[0].set_title(f"Combined Discriminator Loss Mean ")
    ax_combined[0].set_xlabel("Epoch")
    ax_combined[0].set_ylabel("Loss")
    ax_combined[0].legend()

    ax_combined[1].plot(dfs[i]["Epoch"].unique(), [mean[1] for mean in means], label=f"Generator Loss Mean - " + plotNames[i])

    ax_combined[1].set_title(f"Combined Generator Loss Mean ")
    ax_combined[1].set_xlabel("Epoch")
    ax_combined[1].set_ylabel("Loss")
    ax_combined[1].legend()

    ax_combined[2].plot(dfs[i]["Epoch"].unique(), accuracies, label=f"Discriminator Accuracy Mean - " + plotNames[i])

    ax_combined[2].set_title(f"Combined Discriminator Accuracy Mean ")
    ax_combined[2].set_xlabel("Epoch")
    ax_combined[2].set_ylabel("Accuracy")
    ax_combined[2].legend()

plt.tight_layout()
#plt.savefig("/content/drive/MyDrive/Redes neuronales/Monografia/Combined-Training-plot.png")
plt.show()
```
